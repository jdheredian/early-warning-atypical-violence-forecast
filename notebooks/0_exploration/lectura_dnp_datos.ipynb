```python
import pandas as pd
import numpy as np
```

```python
dfs_list = []
```

```python
for i in [1,3,4,5,6,7,8,11,12,13,14,15,16,19,22,26,28,29]:
    dfs_list.append(pd.read_excel(f"/Users/juandiegoheredianino/Downloads/TerriData_Dim{i}.xlsx").groupby(['Dimensión','Subcategoría','Indicador','Unidad de Medida']).agg(
        Año_min=('Año', 'min'),
        Año_max=('Año', 'max'),
        Meses_únicos=('Mes', 'nunique'),
        Entidades_únicas=('Código Entidad', 'nunique')
    ).reset_index())
```

```python
pd.concat(dfs_list, axis=0).reset_index(drop=True)
```
